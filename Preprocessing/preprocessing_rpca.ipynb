```python
audio_path = '/Users/kwang/Desktop/untwist-master/dataset/chuncks_gj2'
save_path = '/Users/kwang/Desktop/untwist-master/rpca/chuncks_gj2'
```

```python
files = [y for y in os.listdir(audio_path) if '.wav' in y]
```

```python
# sample
x = Wave.read(audio_path + '/' + files[3])
X = stft.process(x)
(L, S) = rpca.process(X)
M = RatioMask(np.abs(S), np.abs(L))
v = istft.process(X * M)
v.write(os.path.join(save_path, files[3])) 
```

```python
for i in files:
    x = Wave.read(audio_path + '/' + i)
    X = stft.process(x)    
    (L, S) = rpca.process(X)
    M = RatioMask(np.abs(S), np.abs(L))
    v = istft.process(X * M)
    
    v.write(os.path.join(save_path, i))
```
